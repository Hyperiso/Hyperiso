```python
import sympy as sp

m_b, m_c, m_B, m_D, r_D, w, V_1, S_1 = sp.symbols('m_b m_c m_B m_D r_D w V_1 S_1', positive=True)
```

```python
h_p = 1 / (2 * (1 + r_D ** 2 - 2 * r_D * w)) * (-(1 + r_D) ** 2 * (w - 1) * V_1 + (1 - r_D) ** 2 * (w + 1) * S_1)
h_m = (1 - r_D ** 2) * (w + 1) / (2 * (1 + r_D ** 2 - 2 * r_D * w)) * (S_1 - V_1)
h_T = (m_b + m_c) / (m_B + m_D) * (h_p - (1 + r_D) / (1 - r_D) * h_m)
q2 = m_B ** 2 + m_D ** 2 - 2 * m_B * m_D * w
lam_D = ((m_B - m_D) ** 2 - q2) * ((m_B + m_D) ** 2 - q2)
```

```python
sp.factor(lam_D.subs(m_D, r_D * m_B))
```

```python
F_1 = 1 / (2 * sp.sqrt(m_B * m_D)) * ((m_B + m_D) * h_p - (m_B - m_D) * h_m)
F_1 = sp.factor(F_1.subs(m_D, r_D * m_B))
F_1
```

```python
H_V0 = sp.sqrt(lam_D / q2) * F_1
H_V0 = sp.simplify(sp.factor(H_V0.subs(m_D, r_D * m_B)))
H_V0
```
