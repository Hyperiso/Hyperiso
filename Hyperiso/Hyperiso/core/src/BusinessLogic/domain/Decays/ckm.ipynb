```python
import numpy as np

lam = 0.22501
A = 0.826
rho = 0.1591
eta = 0.3523

s_12 = lam
s_23 = A * lam ** 2
u_13 = A * lam ** 3 * (rho + 1.0j * eta) * np.sqrt(1 - pow(A * lam ** 2, 2)) / (np.sqrt(1 - lam ** 2) * (1. - pow(A * lam ** 2, 2) * (rho + 1.0j * eta)))
s_13 = abs(u_13)
expid = u_13 / abs(u_13)
c_12 = np.sqrt(1 - s_12 * s_12)
c_23 = np.sqrt(1 - s_23 * s_23)
c_13 = np.sqrt(1 - s_13 * s_13)

V_ud = c_12 * c_13
V_us = s_12 * c_13
V_ub = s_13 / expid
V_cd = -s_12 * c_23 - c_12 * s_23 * s_13 * expid
V_cs = c_12 * c_23 - s_12 * s_23 * s_13 * expid
V_cb = s_23 * c_13
V_td = s_12 * s_23 - c_12 * c_23 * s_13 * expid
V_ts = -c_12 * s_23 - s_12 * c_23 * s_13 * expid
V_tb = c_23 * c_13
```

```python
lambda_d = V_cd * np.conj(V_ud)
lambda_s = V_cs * np.conj(V_us)
lambda_b = V_cb * np.conj(V_ub)
lambda_hat_u = V_ub * np.conj(V_us) / V_tb / np.conj(V_ts)
```
