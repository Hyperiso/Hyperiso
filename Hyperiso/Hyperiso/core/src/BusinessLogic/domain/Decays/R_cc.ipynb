```python
from sympy import *

s, m_V, Gamma_tot, Gamma_had, BR, alpha = symbols('s m_V \Gamma_{\mathrm{tot}} \Gamma_{\mathrm{had}} \mathcal{B} \\alpha', real=True, positive=True)
s_0, m_D = symbols('s_0 \hat{m}_D', positive=True)
```

```python
R_cc_res = 9 * s * BR * Gamma_tot * Gamma_had / ((s - m_V ** 2) ** 2 + m_V ** 2 * Gamma_tot ** 2) / alpha
f = R_cc_res / (s * (s - s_0))
```

```python
f
```

```python
s_c = 4 * m_D ** 2
f_0 = 9 * BR * Gamma_tot * Gamma_had / ((s_0 - m_V ** 2) ** 2 + m_V ** 2 * Gamma_tot ** 2) / alpha
f_1 = simplify(apart(f, s).args[0] / f_0)
f_2 = simplify(apart(f, s).args[1] / f_0)
```

```python
L = symbols('\\Lambda', positive=True)
F_1 = simplify(integrate(f_1, (s, s_c, L)))
F_1
```

```python
f_2
F_2 = integrate(f_2, (s, s_c, L))
F_2
```

```python
F_res = limit(F_1 + F_2, L, +oo)
simplify(F_res)
```
