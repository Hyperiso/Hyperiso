```python
from sympy import *
m_B, m_K, q2, tp, tm, t0 = symbols('m_B m_{K^*} q^2 t_+ t_- t_0', real=True) 
A_1, A_12, V, T_1, T_2, T_23 = symbols('A_1 A_{12} V T_1 T_2 T_{23}', real=True) 
```

```python
A_2 = (pow(m_B+m_K,2)*(m_B*m_B-m_K*m_K-q2)*A_1-16*m_B*m_K*m_K*(m_B+m_K)*A_12)/((pow(m_B+m_K,2)-q2)*(pow(m_B-m_K,2)-q2))
```

```python
stm = solve(tp * (1 - sqrt(1 - tm / tp)) - t0, tm)
```

```python
tmB, tmK = solve([(m_B + m_K) ** 2 - tp, (m_B - m_K) ** 2 - stm[0]], (m_B, m_K))[0]
tmK
```

```python
A_2 = factor(A_2.subs({m_B: tmB, m_K: tmK}))
```

```python
factor(expand(A_2).coeff(A_12))
```

```python
a_1, a_2, s_hat, u = symbols("a_1 a_2 \hat{s} u", real=True)

ubar = 1 - u
xi = 2 * u - 1
phi = 6 * u * ubar * (1 + 3 * a_1 * xi + a_2 * (-3 + 15 * xi ** 2) / 2)
f = simplify(phi / (3 * (ubar + s_hat * u)))
```

```python
F_perp = poly(integrate(f, (u, 0, 1)), (a_1, a_2))
factor(F_perp.nth(1, 0))
```
