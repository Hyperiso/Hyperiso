```python
from sympy import *
m_B, m_K, q2, m_b, e = symbols('m_B m_{K^*} q^2 m_b e', real=True) 
s_hat, m_b_hat, m_K_hat = symbols("\hat{s} \hat{m}_b \hat{m}_{K^*}", real=True)
A_1, A_12, V, T_1, T_2, T_23 = symbols('A_1 A_{12} V T_1 T_2 T_{23}', real=True) 
N0 = symbols("N'")
T_perp_p, T_perp_m = symbols('\mathcal{T}_\perp^+ \mathcal{T}_\perp^-')
C7, C7p, C9, C9p = symbols("C_7 C'_7 C_9 C'_9")
```

```python
lam = m_B ** 4 + m_K ** 4 + q2 ** 2 - 2 * (m_B ** 2 * m_K ** 2 + q2 * (m_B ** 2 + m_K ** 2))
E_K = (m_B ** 2 + m_K ** 2 - q2) / (2 * m_B)
xi_perp = m_B / (m_B + m_K) * V
xi_par = (m_B + m_K) / (2 * E_K) * A_1 - (m_B - m_K) / m_B * (pow(m_B+m_K,2)*(m_B*m_B-m_K*m_K-q2)*A_1-16*m_B*m_K*m_K*(m_B+m_K)*A_12)/((pow(m_B+m_K,2)-q2)*(pow(m_B-m_K,2)-q2))
```

```python
simplify(factor(xi_par.subs({q2: 0, m_b: m_B * m_b_hat, m_K: m_B * m_K_hat})))
```

```python
T_tilde_p = (m_B ** 2 - m_K ** 2) / (2 * m_B ** 2) * T_2 - sqrt(lam) / (2 * m_B ** 2) * T_1
T_tilde_m = (m_B ** 2 - m_K ** 2) / (2 * m_B ** 2) * T_2 + sqrt(lam) / (2 * m_B ** 2) * T_1
V_tilde_p = Rational(1, 2) * ((m_B + m_K) / m_B * A_1 - sqrt(lam) / (m_B * (m_B + m_K)) * V)
V_tilde_m = Rational(1, 2) * ((m_B + m_K) / m_B * A_1 + sqrt(lam) / (m_B * (m_B + m_K)) * V)
```

```python
N_p = -1 / (16 * pi ** 2) * m_b / m_B * ((m_B ** 2 - m_K ** 2) * 2 * E_K / m_B ** 3 * T_perp_m - sqrt(lam) / m_B ** 2 * T_perp_p)
N_m = -1 / (16 * pi ** 2) * m_b / m_B * ((m_B ** 2 - m_K ** 2) * 2 * E_K / m_B ** 3 * T_perp_m + sqrt(lam) / m_B ** 2 * T_perp_p)
H_V_p = -I * N0 * (C9 * V_tilde_p - C9p * V_tilde_m + m_B ** 2 / q2 * (2 * m_b / m_B * (C7 * T_tilde_p - C7p * T_tilde_m) - 16 * pi ** 2 * N_p))
H_V_m = -I * N0 * (C9 * V_tilde_m - C9p * V_tilde_p + m_B ** 2 / q2 * (2 * m_b / m_B * (C7 * T_tilde_m - C7p * T_tilde_p) - 16 * pi ** 2 * N_m))
```

```python
A_p = limit(q2 * H_V_p / e, q2, 0)
```

```python
A_p = simplify(factor(A_p.subs({m_b: m_B * m_b_hat, m_K: m_B * m_K_hat}), deep=True))
A_p
```
