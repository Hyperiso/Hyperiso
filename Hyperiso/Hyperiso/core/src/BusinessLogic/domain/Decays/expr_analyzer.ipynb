```python
def extract_coefs(expr, max_s=20, max_ls=10, max_z=5, max_lc=3):
    expr = expr.subs(sqrt(z), sq_z)
    expr = poly(expr, (s, Ls, sq_z, lc))

    kappa = '{\n'
    for i in range(0, max_s + 1):
        for j in range(0, max_ls + 1):
            for l in range(0, 2 * (max_z + 1)):
                for m in range(0, max_lc + 1):
                    a = expr.nth(i, j, l, m).evalf()
                    if a != 0:
                        kappa += '\t{' + str(i) + ', 0, ' + str(l) + ', ' + str(m) + ', ' + str(a) + '},\n'

    return kappa[:-2] + '\n}'
```
