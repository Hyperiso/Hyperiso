```python
from sympy import *

eta4, eta5 = symbols('\\eta_4 \\eta_5', positive=True)
a_i = [Rational(6, 23), Rational(3, 23), Rational(-24, 23), -0.6315, 0.7184]
b_i = [Rational(6, 25), Rational(3, 25), Rational(-24, 25), -0.5810, 0.6610]
```

```python
eta_S_11, eta_S_12, eta_S_21, eta_S_22 = symbols('\\eta_S^{11} \\eta_S^{12} \\eta_S^{21} \\eta_S^{22}') 
```

```python
def format_S(eta):
    # Make eta4 * eta_5 terms explicit
    eta = eta.subs({
        eta4**0.419*eta5**0.3685: eta4 * eta5 * eta_S_11,
        eta4**0.419*eta5**1.7184: eta4 * eta5 * eta_S_12,
        eta4**1.6610*eta5**0.3685: eta4 * eta5 * eta_S_21,
        eta4**1.6610*eta5**1.7184: eta4 * eta5 * eta_S_22
    })
    
    # Make eta4 terms explicit
    eta = eta.subs({
        eta4**0.419*eta5**-0.6315: eta4 * eta_S_11,
        eta4**0.419*eta5**0.7184: eta4 * eta_S_12,
        eta4**1.6610*eta5**-0.6315: eta4 * eta_S_21,
        eta4**1.6610*eta5**0.7184: eta4 * eta_S_22
    })
    
    # Gather remaining monomials
    eta = eta.subs({
        eta4**-0.5810*eta5**-0.6315: eta_S_11,
        eta4**-0.5810*eta5**0.7184: eta_S_12,
        eta4**0.6610*eta5**-0.6315: eta_S_21,
        eta4**0.6610*eta5**0.7184: eta_S_22
    })
    return eta
```

```python
def extract_abc_S(eta):
    eta = expand(format_S(eta))
    eta = poly(eta, (eta_S_11, eta_S_12, eta_S_21, eta_S_22, eta4, eta5))
    a = zeros(2)
    b = zeros(2)
    c = zeros(2)
    a[0, 0] = eta.nth(1, 0, 0, 0, 0, 0)
    a[0, 1] = eta.nth(0, 1, 0, 0, 0, 0)
    a[1, 0] = eta.nth(0, 0, 1, 0, 0, 0)
    a[1, 1] = eta.nth(0, 0, 0, 1, 0, 0)
    b[0, 0] = eta.nth(1, 0, 0, 0, 1, 0)
    b[0, 1] = eta.nth(0, 1, 0, 0, 1, 0)
    b[1, 0] = eta.nth(0, 0, 1, 0, 1, 0)
    b[1, 1] = eta.nth(0, 0, 0, 1, 1, 0)
    c[0, 0] = eta.nth(1, 0, 0, 0, 1, 1)
    c[0, 1] = eta.nth(0, 1, 0, 0, 1, 1)
    c[1, 0] = eta.nth(0, 0, 1, 0, 1, 1)
    c[1, 1] = eta.nth(0, 0, 0, 1, 1, 1)
    
    return a, b, c
```
