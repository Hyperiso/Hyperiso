```python
def format_S(eta):
    # Make eta4_si * eta_5 terms explicit
    eta = eta.subs({
        eta4_si**0.419*eta5_si**0.3685: eta4_si * eta5_si * eta_S_11,
        eta4_si**0.419*eta5_si**1.7184: eta4_si * eta5_si * eta_S_12,
        eta4_si**1.6610*eta5_si**0.3685: eta4_si * eta5_si * eta_S_21,
        eta4_si**1.6610*eta5_si**1.7184: eta4_si * eta5_si * eta_S_22
    })
    
    # Make eta4_si terms explicit
    eta = eta.subs({
        eta4_si**0.419*eta5_si**-0.6315: eta4_si * eta_S_11,
        eta4_si**0.419*eta5_si**0.7184: eta4_si * eta_S_12,
        eta4_si**1.6610*eta5_si**-0.6315: eta4_si * eta_S_21,
        eta4_si**1.6610*eta5_si**0.7184: eta4_si * eta_S_22
    })
    
    # Gather remaining monomials
    eta = eta.subs({
        eta4_si**-0.5810*eta5_si**-0.6315: eta_S_11,
        eta4_si**-0.5810*eta5_si**0.7184: eta_S_12,
        eta4_si**0.6610*eta5_si**-0.6315: eta_S_21,
        eta4_si**0.6610*eta5_si**0.7184: eta_S_22
    })
    return eta
```

```python
def extract_abc_S(eta):
    eta = expand(format_S(eta))
    eta = poly(eta, (eta_S_11, eta_S_12, eta_S_21, eta_S_22, eta4_si, eta5_si))
    a = zeros(2)
    b = zeros(2)
    c = zeros(2)
    a[0, 0] = eta.nth(1, 0, 0, 0, 0, 0)
    a[0, 1] = eta.nth(0, 1, 0, 0, 0, 0)
    a[1, 0] = eta.nth(0, 0, 1, 0, 0, 0)
    a[1, 1] = eta.nth(0, 0, 0, 1, 0, 0)
    b[0, 0] = eta.nth(1, 0, 0, 0, 1, 0)
    b[0, 1] = eta.nth(0, 1, 0, 0, 1, 0)
    b[1, 0] = eta.nth(0, 0, 1, 0, 1, 0)
    b[1, 1] = eta.nth(0, 0, 0, 1, 1, 0)
    c[0, 0] = eta.nth(1, 0, 0, 0, 1, 1)
    c[0, 1] = eta.nth(0, 1, 0, 0, 1, 1)
    c[1, 0] = eta.nth(0, 0, 1, 0, 1, 1)
    c[1, 1] = eta.nth(0, 0, 0, 1, 1, 1)
    
    return a, b, c
```

```python
def r_chi_Bd(m_Bd, m_b, r_bs, r_sl):
    den = m_b * (1 + 1 / (r_bs * r_sl))
    return (m_Bd / den) ** 2
```

```python
import random as rd
import numpy as np

def u(f, p_0: list, u_p: list):
    N = 10000
    evals = []
    for _ in range(N):
        p = [rd.normalvariate(p_, u_) for (p_, u_) in zip(p_0, u_p)]
        evals.append(f(*p))
    return np.std(evals)
```

```python
def alpha_s(mu, lambda_qcd, n_f):
    b0 = (11 - 0.66667 * n_f) / (4 * np.pi)
    b1 = (102 - 12.6667 * n_f) / (4 * np.pi) ** 2
    b2 = (1428.5 - 279.611 * n_f + 6.01852 * n_f ** 2) / (4 * np.pi) ** 3
    b3 = (29243 - 6946.3 * n_f + 405.089 * n_f ** 2 + 1.49931 * n_f ** 3) / (4 * np.pi) ** 4
    b4 = (1024 * (524.56 - 181.8 * n_f + 17.16 * n_f ** 2 - 0.22586 * n_f ** 3 - 0.0017993 * n_f ** 4)) / (4 * np.pi) ** 5
    t = np.log(mu ** 2 / lambda_qcd ** 2)
    l = np.log(t)

    return 1 / (b0 * t) * (
        1 
        - b1 * l / (b0 ** 2 * t)
        + (b1 ** 2 * (l ** 2 - l - 1) + b0 * b2) / (b0 ** 4 * t ** 2)
        + (b1 ** 3 * (-2 * l ** 3 + 5 * l ** 2 + 4 * l - 1) - 6 * b0 * b2 * b1 * l + b0 ** 2 * b3) / (2 * b0 ** 6 * t ** 3)
        + (18 * b0 * b2 * b1 ** 2 * (2 * l ** 2 - l - 1) + b1 ** 4 * (6 * l ** 4 - 26 * l **3 - 9 * l ** 2 + 24 * l + 7) - b0 ** 2 * b3 * b1 * (12 * l + 1) + 2 * b0 ** 2 * (5 * b2 ** 2 + b0 * b4)) / (6 * b0 ** 8 * t ** 4) 
    )
```

```python
g = sum(series(f * a, a, 0, 3).args[:-1]) / a
alpha = symbols('alpha', positive=True)
g = simplify(g.subs(a, alpha / (4 * pi)))
```

```python
alpha_1, alpha_0 = symbols('\\alpha(\\mu_{1}) \\alpha(\\mu_{0})', positive=True)
u = integrate(g, (alpha, alpha_0, alpha_1))
eta = symbols('\\eta', positive=True)
u = u.subs(alpha_1, alpha_0 / eta)
u = u.subs(alpha_0, 4 * pi * a)
u = factor(expand(u))
U = series(exp(u), a, 0, 3)
U = U.args[0] + U.args[1] + U.args[3]
U = U.subs(exp(g0 * log(eta) / (2 * b0)), eta ** (g0 / (2 * b0)))
U = U.subs(exp(-g0 * log(eta) / (2 * b0)), eta ** (-g0 / (2 * b0)))
U = U.subs(sqrt(eta ** (g0 / b0)), eta ** (g0 / (2 * b0)))
U = simplify(U)
```

```python
beta_dict = {b0: nb0, b1: nb1, b2: nb2}
gamma_p_dict = {g0: gp0, g1: gp1, g2: gp2}
gamma_m_dict = {g0: gm0, g1: gm1, g2: gm2}

def eval_p(m, n_f):
    m = m.subs(beta_dict)
    m = m.subs(gamma_p_dict)
    m = m.subs(nf, n_f)
    return simplify(m)

def eval_m(m, n_f):
    m = m.subs(beta_dict)
    m = m.subs(gamma_m_dict)
    m = m.subs(nf, n_f)
    return simplify(m)
```
